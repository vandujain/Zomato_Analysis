```python
import pandas as pd #data manipulation
import seaborn as sns #visualization
import numpy as np #maths + stats
import matplotlib.pyplot as plt #visualization
```

```python
orders = pd.read_csv("C:/Users/vanda/Desktop/py/Orders.csv")
customers = pd.read_csv("C:/Users/vanda/Desktop/py/Customers.csv")
restaurants = pd.read_csv("C:/Users/vanda/Desktop/py/Restaurants.csv")
```

```python
#checking orders info 

orders.info()
```

```python
#checking customers info 

customers.info()
```

```python
#checking restaurants info 

restaurants.info()
```

```python
#Top Cuisines by Order count

top_cuisines = restaurants['Cuisine_Types'].value_counts().head(10)
print(top_cuisines)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='viridis')
plt.title('Top 10 Cuisines Ordered')
plt.xlabel('Order Count')
plt.ylabel('Cuisine')
plt.show()
```

```python
#Orders with respect to time

orders["Order_Date"] = pd.to_datetime(orders["Order_Date"])
orders
```

```python
#Orders with respect to month
orders["Month"] = orders["Order_Date"].dt.strftime("%B")
orders
```

```python
#Orders with respect to weekdays
orders["Weekdays"] = orders["Order_Date"].dt.strftime("%A")
orders
```

```python
orders["Month"] = orders["Order_Date"].dt.strftime("%B")
orders
plt.figure(figsize=(6,3))
sns.countplot(x = "Month", data = orders, hue = "Month",palette = 'viridis')
plt.xticks(rotation = 45)
plt.show()
```

```python
#order count with respect to months
orders["Order_Date"] = pd.to_datetime(orders["Order_Date"])
orders
plt.figure(figsize=(6, 3))
sns.countplot(x="Month", data=orders, color="red", order=["January", "February", "March",
                                                          "April", "May", "June", "July",
                                                          "August", "September", "October"])
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Order Count")
plt.title("Monthly order counts")
plt.show()
```

```python
#Total amount w.r.t time
orders ["Hour"] = orders ["Order_Date"].dt.strftime("%H")
orders ["Hour"] = orders ["Hour"].astype("int")
orders = orders.sort_values (by = "Hour")
sns.lineplot(x = "Hour", y = "Total_Amount", data=orders)
plt.show()
```

```python
#Total amount w.r.t time
orders ["Hour"] = orders ["Order_Date"].dt.strftime("%H")
orders ["Hour"] = orders ["Hour"].astype("int")
orders = orders.sort_values (by = "Hour")
sns.lineplot(x = "Hour", y = "Total_Amount", data=orders,estimator = "sum")
plt.show()
```

```python
# merging customers data with orders data by inner join
df = pd.merge(left = orders , right = customers , on = "Customer_ID", how = "inner")
df
```

```python
#The first plot shows the distribution of food orders across various cities. It appears that Chennai has the highest number of orders, followed by Mumbai and Delhi. 
# Kolkata has the lowest number of orders.
plt.figure(figsize=(12,4))
sns.countplot(x = "Customer_Location", data = df)
plt.show()
```

```python
# The second plot breaks down the order status by city. 
# We can see that Completed orders dominate in all cities. However, Cancelled orders are more prevalent in Mumbai and Delhi compared to other cities.
plt.figure(figsize=(10,4))
sns.countplot(x = "Customer_Location", data = df,hue = "Order_Status")
plt.show()
```

```python
#the distribution of food orders across various cities.Each city has a unique preference for certain dishes.Some dishes are more popular across multiple cities.
plt.figure(figsize=(10,5))   #width,height
sns.countplot(x = "Customer_Location", data = df,hue = "Dish_Name",palette = "plasma",edgecolor = "black")
plt.legend(bbox_to_anchor = (1,1))
plt.show()
Dish-wise Insights:

#Garlic Naan, Chole Bhature, and Paneer Tikka: Popular across multiple cities, especially Mumbai and Delhi.
#Mutton Rogan Josh, Chicken Biryani, and Masala Dosa: Popular in Chennai, Hyderabad, and Kolkata.
#Veg Manchurian and Egg Fried Rice: Popular across multiple cities, especially Delhi, Bangalore, and Pune.
#Fish Curry, Chicken Shawarma, and Chicken Wings: Popular in Kolkata.
#Butter Chicken and Paneer Butter Masala: Less popular overall, with some demand in Mumbai and Delhi.
```

```python
# Plotting Customer Location Counts by Age Group
plt.figure(figsize=(10,5))
sns.countplot(x = "Customer_Location", data = df,hue = "Customer_Age_Group",palette = "viridis",edgecolor = "black")
plt.legend(bbox_to_anchor = (1,1))
plt.title("Customer Location Distribution by Age Group")
plt.xticks(rotation=20)  
plt.show()
```

```python
#Customer Insights 
#By plotting top 10 customers by order count

top_customers = customers['Customer_Name'].value_counts().head(10)
print(top_customers)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='flare')
plt.title('Top 10 Customers by Order Count')
plt.xlabel('Order Count')
plt.ylabel('Customer Name')
plt.show()
```

```python
#Top 10 Most Popular Restaurants

top_restaurants = restaurants['Name'].value_counts().head(10)
print(top_restaurants)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_restaurants.values, y=top_restaurants.index, palette='plasma')
plt.title('Top 10 Most Popular Restaurants')
plt.xlabel('Number of Orders')
plt.ylabel('Restaurant Name')
plt.show()
```

```python
#Average Ratings of Top 10 Restaurants

top_restaurant_ratings = restaurants.groupby('Name')['Ratings'].mean().nlargest(10)
print(top_restaurant_ratings)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_restaurant_ratings.values, y=top_restaurant_ratings.index, palette='coolwarm')
plt.title('Top 10 Restaurants by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant Name')
plt.show()
```

```python
# merging customers data with orders data by inner join
df = pd.merge(left=orders, right=restaurants, on="Restaurant_ID", how="inner")
df
```

```python
#The plot shows the distribution of ratings on a scale of 3.0 to 5.0. It appears to represent customer reviews.
plt.figure(figsize=(12,4))
sns.countplot(x="Ratings", data=df)
plt.show()
```

```python
# The plot shows the distribution of payment methods used by customers.
plt.figure(figsize=(10,3))
sns.countplot(x="Payment_Method", data=df)
plt.show()
```

```python
#This plot shows the number of completed and canceled orders across different locations.
plt.figure(figsize=(10,4))
sns.countplot(x = "Location", data = df,hue = "Order_Status")
plt.show()
```

```python
# This plot shows the distribution of review counts for different locations. 
plt.figure(figsize=(10,5))
sns.countplot(x = "Location", data = df,hue = "Reviews_Count")
plt.show()
```

```python
#This plot shows the distribution of popular dishes across different locations. Here are the key insights:
#Paneer Tikka: This dish is the most popular across all locations, indicating a wide appeal.
#Chole Bhature: This dish is also quite popular, especially in Mumbai and Delhi.
plt.figure(figsize=(10,5))   #width,height
sns.countplot(x = "Location", data = df,hue = "Dish_Name",palette = "plasma",edgecolor = "purple")
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

```python
#This plot shows the distribution of different cuisine types across locations. Here are the key insights:
#Indian Cuisine: This is the most popular cuisine type across all locations, indicating a strong preference for local cuisine.
#Continental Cuisine: This cuisine type is also popular, especially in Bangalore and Delhi.
plt.figure(figsize=(10,5))   #width,height
sns.countplot(x = "Location", data = df,hue = "Cuisine_Types",palette = "inferno",edgecolor = "green")
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

```python
#The plot shows the distribution of ratings (3.2, 3.6, 4.0, 4.4, 4.8) across different locations.
plt.figure(figsize=(10,5))
sns.countplot(x = "Location", data = df,hue = "Ratings",palette = "coolwarm",edgecolor = "black")
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```
