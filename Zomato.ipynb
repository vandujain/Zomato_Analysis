```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load all datasets
customers = pd.read_csv("Customers.csv")
orders = pd.read_csv("Orders.csv")
restaurants = pd.read_csv("Restaurants.csv")

# Step 2: Merge the datasets
# Assume Orders has 'CustomerID' and 'RestaurantID' for joining
combined_df = orders.merge(customers, on='Customer_ID', how='left') \
                    .merge(restaurants, on='Restaurant_ID', how='left')

# Step 3: Select numeric columns and clean data
numeric_df = combined_df.select_dtypes(include=['float64', 'int64']).dropna()

# Optional: Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# Step 4: Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
combined_df['Cluster'] = kmeans.fit_predict(scaled_data)

# Step 5: Visualize Clusters (using 2 principal components if needed)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(scaled_data)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=reduced[:, 0], y=reduced[:, 1], hue=combined_df['Cluster'], palette='plasma')
plt.title('Customer Segmentation using K-Means')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

# Interpretation of the K-Means customer segmentation graph:
#### Five distinct customer segments were identified using K-Means, each represented by a different color.
#### Cluster 0 (dark blue) is more compact and separate, possibly indicating high-value or loyal customers.
#### Clusters 13 show overlap, suggesting similar average behavior among those customer groups.
#### Cluster 4 (yellow) is widely spread, which may reflect inconsistent or one-time customer behavior.
#### The use of PCA helps visualize complex data in 2D, revealing natural groupings useful for targeted marketing or service optimization.
